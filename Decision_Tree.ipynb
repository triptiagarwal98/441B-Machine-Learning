# 2.) Choosing 4 features to predict the visa approval status. Case_status is the target variable.


```python
#choice is based on whether at least 1/3rd of the data is non-null values. I have also chosen features in such a way to include categorical
#and continuous variables
df_select = df[["class_of_admission", "job_info_experience_num_months","job_info_work_state","country_of_citizenship","case_status"]]
```

# 3.) Cleaning the data for a decision tree.


```python
df_select = df_select.dropna()
```

```python
df_select.info()
```

```python
df_select
```

```python
predictors = df_select[['job_info_experience_num_months']]
```

```python
#creating dummies for string variables
dummies = pd.get_dummies(df_select[["class_of_admission", "country_of_citizenship", "job_info_work_state"]])
x = pd.concat([dummies, predictors], axis=1)
y = df_select["case_status"]
```

# 4.) Fitting and plotting a decision tree of depth 3.


```python
from sklearn import tree
```

```python
clf = tree.DecisionTreeClassifier(max_depth = 3)
clf.fit(x,y)
```

```python
plt.figure(figsize = (20,8))
tree.plot_tree(clf, feature_names = x.columns)
```

```python
y.unique()
```

# 5.) Interpreting on the basis of the largest leaf node (by sample size).


On the basis of largest sample size, this decision tree determines that the Job State - Texas is the most important variable in predicting case status. So when we start from the top of the node we have the following decision pattern: 
1. If the person's job state in Texas = 0, this implies that the person works in any state except Texas.
2. From the above sub-category, on the basis of larger sample size, we check whether the person works in New Jersey or not (following the left node). The smaller node checks if the person, if working in Texas, is a citizen of India or not. But we will follow the larger sample size node of New Jersey.
3. Again the larger sample is the subset of people NOT working in New Jersey. We stop at 3 layers as that is the defined depth.

Finally, based on largest sample sizes, we reach a subset of people that work in a state other than Texas and New Jersey. Here, we find that 52145 are Denied, 41834 are Certified, 4132 are Certified-Expired, and 3348 are Withdrawn. Hence, most people in this subset are predicted to have a Denied case status.

# 6.) Using a for loop to determine the best max_depth for out-of sample accuracy based on training and testing data.

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, train_size = 0.8, random_state = 425)
```

```python
from sklearn import metrics
```

```python
max_depths = [2,3,5]

for md in max_depths:
  clf = tree.DecisionTreeClassifier(max_depth = md)
  clf.fit(x_train,y_train)
  y_pred = clf.predict(x_test)
  print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
```

Hence, the best max depth is 5 as it has the highest classification accuracy of 54%!!
